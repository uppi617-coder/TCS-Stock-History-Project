```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Date', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Date")
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Open', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Open")
plt.xlabel('Open')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'High', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of High")
plt.xlabel('High')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Low', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Low")
plt.xlabel('Low')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Close', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Close")
plt.xlabel('Close')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Dividends', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Dividends")
plt.xlabel('Dividends')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Stock Splits', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Year")
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Open', color = 'blue',
             edgecolor = 'linen', alpha = 0.5, bins = 5)

plt.title("Histogram of Open")
plt.xlabel('Open')
plt.ylabel('Count')
plt.vlines(data['Open'].mean(), ymin = 0, ymax = 4000, colors='blue', label='Mean')
plt.vlines(data['Open'].median(), ymin = 0, ymax = 4000, colors='red', label='Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Dividends', color = 'blue',
             edgecolor = 'linen', alpha = 0.5, bins = 5)

plt.title("Histogram of Dividends")
plt.xlabel('Dividends')
plt.ylabel('Count')
plt.vlines(data['Dividends'].mean(), ymin = 0, ymax = 4000, colors='blue', label='Mean')
plt.vlines(data['Dividends'].median(), ymin = 0, ymax = 4000, colors='red', label='Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.distplot(data['Open'], color = 'blue')

plt.title("Distplot of Open")
plt.xlabel('Open')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.distplot(data['Dividends'], color = 'blue')

plt.title("Distplot of Dividends")
plt.xlabel('Dividends')
plt.ylabel('Density')
plt.show()
```
